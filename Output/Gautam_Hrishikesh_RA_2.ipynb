```python
world_merged=pd.merge(aqli_shp,gadm_aqli,on=["name2"] )
```

```python
western_eu_data=world_merged.loc[world_merged['name0'].isin(Western_Europe)]
```

```python
# Plotting Potential gain in life expectancy (relative to the WHO guideline) in western europe in AQLI map Color scheme

# Define the colormap and the bins
colors = ['#ffffff', '#fff5eb', '#fee6ce', '#fdd0a2', '#fdae6b', 
          '#fd8d3c', '#f16913', '#d94801', '#a63603', '#7f2704']
bounds = [0, 0.1, 0.5, 1, 2, 3, 4, 5, 6, 6.0]
cmap = mcolors.LinearSegmentedColormap.from_list('CustomMap', colors)
norm = mcolors.BoundaryNorm(bounds, cmap.N)

fig, ax = plt.subplots(figsize=(15, 10))
western_eu_data.plot(column='llpp_who_2021', ax=ax, legend=True,
                     legend_kwds={'label': "Potential gain in life expectancy (relative to the WHO guideline)"},
                     cmap='OrRd', missing_kwds={'color': 'lightgrey'},norm=norm)

ax.set_axis_off()
```

```python
# Plotting Potential gain in life expectancy (relative to the WHO guideline) in eastern europe in AQLI map Color scheme
Eastern_eu_data=world_merged.loc[world_merged['name0'].isin(Eastern_Europe)]
colors = ['#ffffff', '#fff5eb', '#fee6ce', '#fdd0a2', '#fdae6b', 
          '#fd8d3c', '#f16913', '#d94801', '#a63603', '#7f2704']
bounds = [0, 0.1, 0.5, 1, 2, 3, 4, 5, 6, 6.0]
cmap = mcolors.LinearSegmentedColormap.from_list('CustomMap', colors)
norm = mcolors.BoundaryNorm(bounds, cmap.N)
# Plotting
fig, ax = plt.subplots(figsize=(15, 10))
Eastern_eu_data.plot(column='llpp_who_2021', ax=ax, legend=True,
                     legend_kwds={'label': "Potential gain in life expectancy (relative to the WHO guideline)"},
                     cmap='OrRd', missing_kwds={'color': 'lightgrey'},norm=norm)

ax.set_axis_off()

# Save the map
plt.savefig('world_map_with_llpp.png', format='png', dpi=300)
```

```python
#A part of Chukot region in easternmost russia is being plotted in the left corner of the plot near Alaska and shifting the whole plot towards right
#So, from here onwards I am dropping Chukot level1 region from eastern europe map so that it can be visualised properly

Eastern_eu_data=world_merged.loc[world_merged['name0'].isin(Eastern_Europe)]
Eastern_eu_dropped = Eastern_eu_data[Eastern_eu_data['name1'] != 'Chukot']
# Plotting
colors = ['#ffffff', '#fff5eb', '#fee6ce', '#fdd0a2', '#fdae6b', 
          '#fd8d3c', '#f16913', '#d94801', '#a63603', '#7f2704']
bounds = [0, 0.1, 0.5, 1, 2, 3, 4, 5, 6, 6.0]
cmap = mcolors.LinearSegmentedColormap.from_list('CustomMap', colors)
norm = mcolors.BoundaryNorm(bounds, cmap.N)
fig, ax = plt.subplots(figsize=(15, 10))
Eastern_eu_dropped.plot(column='llpp_who_2021', ax=ax, legend=True,
                     legend_kwds={'label': "Potential gain in life expectancy (relative to the WHO guideline)"},
                     cmap='OrRd', missing_kwds={'color': 'lightgrey'},norm=norm)

ax.set_axis_off()
ax.set_aspect('auto')
```

```python
# Comparing Potential gain in life expectancy (relative to the WHO guideline) in western vs eastern europe in AQLI map Color scheme
fig, axs = plt.subplots(1, 2,figsize=(12, 6))
fig.suptitle('Potential gain in life expectancy in Europe', fontsize=16,weight="bold")

colors = ['#ffffff', '#fff5eb', '#fee6ce', '#fdd0a2', '#fdae6b', 
          '#fd8d3c', '#f16913', '#d94801', '#a63603', '#7f2704']
bounds = [0, 0.1, 0.5, 1, 2, 3, 4, 5, 6, 6.0]
cmap = mcolors.LinearSegmentedColormap.from_list('CustomMap', colors)
norm = mcolors.BoundaryNorm(bounds, cmap.N)

western_eu_data.plot(column='llpp_who_2021', ax=axs[0], legend=False,
                     legend_kwds={'label': "Potential gain in life expectancy (relative to the WHO guideline)"},
                     cmap='OrRd', missing_kwds={'color': 'lightgrey'},norm=norm)

axs[0].set_axis_off()
axs[0].set_aspect('auto')
axs[0].set_title('Western Europe',weight="bold", fontsize=14)
Eastern_eu_dropped.plot(column='llpp_who_2021', ax=axs[1], legend=True,
                     legend_kwds={'label': "Potential gain in life expectancy (relative to the WHO guideline)"},
                     cmap='OrRd', missing_kwds={'color': 'lightgrey'},norm=norm)

axs[1].set_axis_off()
axs[1].set_aspect('auto')
axs[1].set_title('Eastern Europe',weight="bold", fontsize=14)
fig.savefig('Potential gain in life expectancy in Europe.png', dpi=300)
```
